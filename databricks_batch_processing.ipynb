```
joined2_df = df_geo.join(df_pin, df_geo["ind"] == df_pin["ind"], how="inner")
display(joined2_df)
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import avg

# Creating a Window specification
window_spec = Window.partitionBy("country").orderBy("poster_name")

# Applying window function to calculate average salary within each gender partition
result_df = joined2_df.withColumn("follower_count", count("follower_count").over(window_spec))

# Selecting only the relevant columns
result_df = result_df.select("country", "poster_name", "follower_count").distinct()

display(result_df)
```

```
user_country_most_followers = result_df.select('country', 'follower_count')\
    .orderBy('follower_count', ascending=False)
display(user_country_most_followers.limit(1))
```

```
task7_df = df_user.join(df_pin, df_user["ind"] == df_pin["ind"], how="inner")
display(task7_df)
```

```
age_demongraphics_categorical = task7_df.withColumn('age_group', 
                  when(task7_df.age.between(18, 24), lit('18-24')).otherwise(
                      when(task7_df.age.between(25, 35), lit('25-35')).otherwise(
                          when(task7_df.age.between(36, 50), lit('36-50')).otherwise(lit('50+'))
                      )
                  )
).groupBy('age_group', 'category').agg(count('*').alias('category_count'))
display(age_demongraphics_categorical)
```

```
import pyspark.sql
from pyspark.sql import Window
window = pyspark.sql.Window.partitionBy('age_group').orderBy(desc('category_count'))
ranked = age_demongraphics_categorical.withColumn('rank', row_number().over(window))
top_category_by_age = ranked.filter(col('rank')==1).select('age_group', 'category', 'category_count')
display(top_category_by_age)
```

```
age_demongraphics_categorical = task8_df.withColumn('age_group', 
                  when(task8_df.age.between(18, 24), lit('18-24')).otherwise(
                      when(task8_df.age.between(25, 35), lit('25-35')).otherwise(
                          when(task8_df.age.between(36, 50), lit('36-50')).otherwise(lit('50+'))
                      )
                  )
).groupBy('age_group').agg(percentile_approx('follower_count', 0.5).alias('median_follower_count')).orderBy('age_group')
     
display(age_demongraphics_categorical)
```
